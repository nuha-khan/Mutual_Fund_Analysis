# Live NAV Data Extraction

This notebook demonstrates how live Net Asset Value (NAV) data can be retrieved from the mfapi.in REST API. The extracted data provides the most recent NAV history for selected mutual fund schemes and can be used to supplement historical datasets during analysis.

## Import Required Libraries

Import the Python libraries required for sending API requests and organizing the retrieved JSON response into a tabular format.

```python
import requests
import pandas as pd
```

## Single Scheme API Demonstration

This section demonstrates how live NAV data can be retrieved for a single mutual fund scheme using its AMFI code. It illustrates the API structure and the JSON response before extending the process to multiple schemes.

```python
response = requests.get("https://api.mfapi.in/mf/125497")
data = response.json()
# print(data["meta"])

nav_data = data["data"]
df = pd.DataFrame(nav_data)

df.to_csv("data/raw/live_nav_mfapi.csv",index=False)
print("Live NAV data saved as raw CSV successfully!")
print(df.head())
```

## Specify Mutual Fund Scheme Codes

A list of AMFI scheme codes is defined for the mutual fund schemes whose live NAV information will be retrieved from the API.

```python
scheme_codes = [119551,120503,118632,119092,120841]
```

## Fetch Live NAV Data

For each AMFI scheme code, an API request is sent to mfapi.in. The JSON response is parsed to extract the scheme name and historical NAV records, which are then converted into a Pandas DataFrame for inspection.

```python
for code in scheme_codes:
    url = f"https://api.mfapi.in/mf/{code}"

    response = requests.get(url)
    data = response.json()

    scheme_name = data["meta"]["scheme_name"]
    print("AMFI Code - ",code,"Scheme Name - ",scheme_name)
    nav_data = data["data"]
    df = pd.DataFrame(nav_data)
    print(df.head())
```

## Conclusion

The notebook successfully retrieves live NAV information from the mfapi.in API for one or more mutual fund schemes. The extracted data can be integrated into the ETL workflow or used independently for validating and analyzing historical NAV trends.
